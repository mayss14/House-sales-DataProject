# About the Dataset

This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015. It was taken fromÂ [here](https://www.kaggle.com/harlfoxem/housesalesprediction?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-wwwcourseraorg-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDA0101ENSkillsNetwork20235326-2022-01-01).

| Variable      | Description                                                                                                 |
| ------------- | ----------------------------------------------------------------------------------------------------------- |
| id            | A notation for a house                                                                                      |
| date          | Date house was sold                                                                                         |
| price         | Price is prediction target                                                                                  |
| bedrooms      | Number of bedrooms                                                                                          |
| bathrooms     | Number of bathrooms                                                                                         |
| sqft_living   | Square footage of the home                                                                                  |
| sqft_lot      | Square footage of the lot                                                                                   |
| floors        | Total floors (levels) in house                                                                              |
| waterfront    | House which has a view to a waterfront                                                                      |

| grade         | overall grade given to the housing unit, based on King County grading system                                |

| sqft_basement | Square footage of the basement                                                                              |
| yr_built      | Built Year                                                                                                  |
| yr_renovated  | Year when house was renovated                                                                               |
| zipcode       | Zip code                                                                                                    |
| lat           | Latitude coordinate                                                                                         |
| long          | Longitude coordinate                                                                                        |
| sqft_living15 | Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area |

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
filepath='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/kc_house_data_NaN.csv'
```

```python
df = pd.read_csv(filepath)
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

We delete the columns ID and Unnamed

```python
df.drop(['id','Unnamed: 0'],inplace = True,axis = 1)
df
```

We can see we have missing values for the columns <code> bedrooms</code>  and <code> bathrooms </code>


```python
df['bedrooms'].isnull().sum(), df['bathrooms'].isnull().sum()
```

We can replace the missing values of the column <code>'bedrooms'</code> with the mean of the column  <code>'bedrooms' </code> using the method <code>replace()</code>.


```python
df.replace(np.nan,df['bedrooms'].mean(),inplace = True)
df['bedrooms'].isnull().sum()
```

We also replace the missing values of the column <code>'bathrooms'</code> with the mean of the column  <code>'bathrooms' </code> using the method <code>replace()</code>.

```python
df.replace(np.nan,df['bathrooms'].mean(),inplace= True)
df['bathrooms'].isnull().sum()
```

```python
df['floors'].value_counts().to_frame()
```

We can notice that most of houses have only 1 floor

## Model Development


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
```

```python
X = df[features]
Y = df['price']

model = LinearRegression()
model.fit(X,Y)
```

Calculate the value of R2 to evaluate the performance of our regression model.

```python
from sklearn.metrics import r2_score
y_pred= model.predict(X)
r2_score(Y,y_pred)
```

Now we create a pipeline to scale the data and run the model

```python
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
```

```python
pipeline = Pipeline(Input)
pipeline.fit(X,Y)
```

```python
y_pred= pipeline.predict(X)
```

```python
r2_score(Y,y_pred)
```

The R2 we had is high which proves that our model is a good fit.

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
```

We will now split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)
```

We will use Ridge Model to prevent that our model will be overfitting

```python
from sklearn.linear_model import Ridge
```

```python
model2 = Ridge(alpha=0.1)
model2.fit(x_train,y_train)
```

```python
y_pred = model2.predict(x_test)
r2_score(y_test,y_pred)
```
